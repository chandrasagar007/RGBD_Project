```python
for i in os.listdir(os.path.join('/media/sagar/HDD1/IIIT_PGSSP/CV/Project/python/eccv14-data/benchmarkData/groundTruth')):
    annots = loadmat(os.path.join('/media/sagar/HDD1/IIIT_PGSSP/CV/Project/python/eccv14-data/benchmarkData/groundTruth',str(i)))
    x = np.array(annots['groundTruth'])[0][0][0][0][0]
    y = np.array(annots['groundTruth'])[0][0][0][0][1]
    z = np.array(annots['groundTruth'])[0][0][0][0][2]
#     print(x)
    annots_image = np.zeros((425,560,3), 'uint8')
    annots_image[:,:,0] = x
    annots_image[:,:,1] = y
    annots_image[:,:,2] = z
#     print(annots_image)
#     annots_image.save(os.path.join('/media/sagar/HDD1/IIIT_PGSSP/CV/Project/python/eccv14-data/benchmarkData/annotated_images'+".jpg")) 
    plt.imsave(os.path.join('/media/sagar/HDD1/IIIT_PGSSP/CV/Project/python/eccv14-data/benchmarkData/annotated_images/'+str(i[:-4])+'.png'),annots_image)
```
